```python
import matplotlib.pyplot as plt
import numpy as np

num_simulations = 10000

points = np.random.uniform(0, 1, size=(num_simulations, 2))

x = np.arange(0, 1, 0.01)
y = np.arange(0, 1, 0.01)

# Calculate if point is inside the circle
def distance(a, b):
    return(a**2 + b**2)

counts = 0

for i in range(num_simulations):
    r = distance(points[i, 0], points[i, 1])
    
    if r < 1: 
        counts += 1

pi = (4*counts)/num_simulations

X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.scatter(points[:,0], points[:,1], s=0.1)
plt.contour(X, Y, Z, [1], cmap='plasma')
plt.suptitle('Monte Carlo Simulation to Estimate $\pi$')
plt.title(f'N: {num_simulations}, $\pi$ : {pi}', fontsize=10)
plt.margins(0,0)
plt.axis('scaled')
plt.show()
```
